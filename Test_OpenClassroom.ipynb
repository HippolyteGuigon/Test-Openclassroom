```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 

def factorielle(n):
    if n == 0:
        return 1
    else:
        return n * factorielle(n-1)
```

```python
liste_fact = [factorielle(i) for i in range(1, 100)]
```

```python
plt.figu
plt.plot(liste_fact, "o");
```
